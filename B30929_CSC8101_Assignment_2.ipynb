Q1: Electricity Billing Optimization. You are analyzing monthly household electricity consumption data for 10 customers in
Kampala. Each record contains: customer_name, units_used, connection_type (e.g., 'Bosco',
245, 'Domestic').

```python
customer_data = [
    ('Bosco', 245, 'Domestic'),
    ('MTN', 400, 'Commercial'),
    ('Abul', 520, 'Domestic'),
    ('Isaac', 310, 'Domestic'),
    ('DTB', 890, 'Commercial'),
    ('Grace', 610, 'Domestic'),
    ('Madivani', 150, 'Commercial'),
    ('Mary', 720, 'Domestic'),
    ('Airtel', 480, 'Commercial'),
    ('Zia', 555, 'Domestic')
]
```

1.1. Use zip() and map() to pair customer names with total bills computed by a lambda:
 • Domestic rate = 820 UGX/unit
 • Commercial rate = 1000 UGX/unit

```python
bill_calc = lambda data: data[1] * (820 if data[2] == 'Domestic' else 1000)
customer_names = [c[0] for c in customer_data]
customer_bills = list(map(bill_calc, customer_data))

paired_bills = list(zip(customer_names, customer_bills))

```

1.2. Use filter() to separate high-consumption customers (units > 500).

```python
high_consumers = list(filter(lambda customer: customer[1] > 500, customer_data))
```

1.3. Using *args, write a function average_bill(*bills) that computes the average.

```python
def average_bill(*customer_bills):
    return sum(customer_bills) / len(customer_bills) if customer_bills else 0

customer_avg_bill = average_bill(*customer_bills)
```

1.4. Display results using list comprehension in the format: ['Bosco: UGX 200,900', 'Jane:
UGX 400,000', ...]

```python
display_list = [name + ": UGX " + str(bill) + ":," for name, bill in paired_bills]

print("Customer Bills:")
print(display_list)
print("High Consumption Customers (units > 500):")
print([customer[0] for customer in high_consumers])
print("Average Bill: UGX ", customer_avg_bill)
```

Q2: Market Basket Price Aggregator
Each vendor reports different fruit prices daily:
mango_prices = [2500, 2700, 2600, 2800]
orange_prices = [3000, 3200, 3100, 3050]
apple_prices = [4500, 4600, 4550, 4700]
Attempt the following tasks:


```python
from itertools import zip_longest

mango_prices = [2500, 2700, 2600, 2800]
orange_prices = [3000, 3200, 3100, 3050]
apple_prices = [4500, 4600, 4550, 4700]
```

2.1 Use zip_longest() to handle missing prices.

```python
combined_prices = list(zip_longest(mango_prices, orange_prices, apple_prices, fillvalue=0))
#print(combined_prices)
print("Combined Daily Prices (Mango, Orange, Apple):")
for day, prices in enumerate(combined_prices, start=1):
    print("Day ", str(day), ": ", prices)
```

2.2 Define a lambda function avg = lambda lst: sum(lst)/len(lst) to compute average per fruit.

```python
get_avg_price = lambda lst: sum(lst) / len(lst)
avg_price_mango = get_avg_price(mango_prices)
avg_price_orange = get_avg_price(orange_prices)
avg_price_apple = get_avg_price(apple_prices)

print("Average Prices:")
print("Mango: ", str(avg_price_mango),", Orange: ", avg_price_orange, ", Apple: ", str(avg_price_apple))
```

2.3 Create a generator expression that yields all fruits with average price above 3000.

```python
fruit_avgs = {
    "Mango": avg_price_mango,
    "Orange": avg_price_orange,
    "Apple": avg_price_apple
}

above_avg_fruits = (fruit for fruit, price in fruit_avgs.items() if price > 3000)
```

2.4 Print the generator’s results neatly as: Fruits Above Average Price: Mango, Orange, Apple

```python
print("Fruits Above Average Price:", (", ".join(above_avg_fruits)))
```

kampala = [28, 30, 29, 27, 26, 29, 30, 31, 32, 30, 29, 28]
gulu = [25, 26, 27, 27, 28, 29, 30, 30, 29, 27, 26, 25]
mbarara = [22, 23, 23, 24, 25, 25, 26, 27, 27, 26, 24, 23]
Attempt the following tasks:

```python
kampala = [28, 30, 29, 27, 26, 29, 30, 31, 32, 30, 29, 28]
gulu = [25, 26, 27, 27, 28, 29, 30, 30, 29, 27, 26, 25]
mbarara = [22, 23, 23, 24, 25, 25, 26, 27, 27, 26, 24, 23]
```

3.1. Use map() with lambda to convert all temperatures to Fahrenheit.

```python
# Formula: F = (C * 9/5) + 32
get_fahrenheit = lambda cel: (cel * 9/5) + 32

kampala_F = list(map(get_fahrenheit, kampala))
gulu_F = list(map(get_fahrenheit, gulu))
mbarara_F = list(map(get_fahrenheit, mbarara))

print("Kampala Temperatures (°F): ", kampala_F)
print("Gulu Temperatures (°F): ", gulu_F)
print("Mbarara Temperatures (°F): ", mbarara_F)
```

3.2. Use zip() to pair months (Jan–Dec) with Kampala temps.

```python
months = ["January", "February", "March", "April", "May", "June",
          "July", "August", "September", "October", "November", "December"]

kampala_monthly = list(zip(months, kampala))
gulu_monthly = list(zip(months, gulu))
mbarara_monthly = list(zip(months, mbarara))
print("Kampala Monthly Temperatures (°C): ", kampala_monthly)
print("Gulu Monthly Temperatures (°C): ", gulu_monthly)
print("Mbarara Monthly Temperatures (°C): ", mbarara_monthly)
```

3.3. Use filter() to identify months with temperature > 30°C.

```python
hot_months_kampala = list(filter(lambda m: m[1] > 30, kampala_monthly))
hot_months_gulu = list(filter(lambda m: m[1] > 30, gulu_monthly))
hot_months_mbarara = list(filter(lambda m: m[1] > 30, mbarara_monthly))
print("Kampala Hot Month Temperatures (°C): ", hot_months_kampala)
print("Gulu Hot Month Temperatures (°C): ", hot_months_gulu)
print("Mbarara Hot Month Temperatures (°C): ", hot_months_mbarara)
```

3.4. Create a generator to yield month names and temperatures above threshold.

```python
hot_months_gen = (month + " - " + str(temp) +"°C" for month, temp in hot_months_kampala)
```

3.5. Print as: Hot months in Kampala: July - 31°C, August - 32°C

```python
print("Hot months in Kampala:", ", ".join(hot_months_gen))
```

Q4: School Fees Payment Analyzer
Installment data:
students = ['Alex', 'Grace', 'Sarah', 'Brian']
installments = [
 [150000, 200000, 250000],
 [500000, 0, 200000],
 [300000, 300000, 300000],
 [400000, 100000, 0]
]
Attempt the following tasks:

```python
students = ['Alex', 'Grace', 'Sarah', 'Brian']
installments = [
    [150000, 200000, 250000],
    [500000, 0, 200000],
    [300000, 300000, 300000],
    [400000, 100000, 0]
]
```

4.1. Write a lambda function that sums valid payments and ignores zeros.

```python
sum_valid_payments = lambda pays: sum(x for x in pays if x > 0)
```

4.2. Use map() to compute total paid by each student.

```python
totals = list(map(sum_valid_payments, installments))
```

4.3. Combine student names and totals using zip().

```python
payment_records = list(zip(students, totals))
print("Student Payment Records")
for name, total in payment_records:
    print(name + ": UGX " + str(total), end=", ")
```

4.4. Use filter() to find students who have cleared full fees (≥600,000).

```python
cleared_students = list(filter(lambda rec: rec[1] >= 600000, payment_records))
print("Students Who Cleared Full Fees:")
for name, total in cleared_students:
    print(name + ": UGX " + str(total), end=", ")
```

4.5. Create payment_summary(**kwargs) that receives named args like Alex=600000 and prints summaries.


```python

def payment_summary(**kwargs):
    print("--- Payment Summary ---")
    for name, amount in kwargs.items():
        print(name + ": UGX " + str(amount), end=", ")

payment_summary(**dict(payment_records))
```

Q5: Agricultural Yield Estimator
Yield data:
districts = ['Bushenyi', 'Mityana', 'Kasese', 'Mbale']
yield_data = [1200, 1500, 900, 1300]
Attempt the following tasks:

```python
districts = ['Bushenyi', 'Mityana', 'Kasese', 'Mbale']
yield_data = [1200, 1500, 900, 1300]  # in kilograms
```

5.1. Use a list comprehension with a lambda to convert yields from kg to tons.

```python
kg_to_tons = lambda x: x / 1000 # i ton is equal to 100kg
yields_in_tons = [kg_to_tons(y) for y in yield_data]
print("Yields in tons:", yields_in_tons)
```

5.2. Write a generator that yields only districts with yield > 1 ton.

```python
def high_yield_districts():
    for d, y in zip(districts, yields_in_tons):
        if y > 1:
            yield (d, y)

print("Districts with yield > 1 ton:")
for d, y in high_yield_districts():
    print(d + ": " + str(y) + " tons")
```

5.3. Use *args to compute average yield across all districts.

```python
def average_yield(*args):
    return sum(args) / len(args)

avg = average_yield(*yields_in_tons)
print("Average yield across all districts: " + str(avg) + " tons")
```

5.4. Use **kwargs to simulate price changes per district and compute revenue (tons × price).

```python
def compute_revenue(**kwargs):
    print("\nRevenue per district:")
    for district, price in kwargs.items():
        index = districts.index(district)
        tons = yields_in_tons[index]
        revenue = tons * price
        print(district, " produced ", str(tons), " tons — Revenue: UGX ", str(revenue))

price_data = {
    'Bushenyi': 4_000_000,
    'Mityana': 4_500_000,
    'Kasese': 5_000_000,
    'Mbale': 4_200_000
}

compute_revenue(**price_data)
```

Q6: Web Data Aggregation
Given that sites = ['https://ucu.ac.ug', 'https://harba.ug', 'https://www.bou.or.ug']
Attempt the following tasks:

```python
import requests
sites = ['https://ucu.ac.ug', 'https://harba.ug', 'https://www.bou.or.ug']
```

6.1. Write a function that takes *urls and uses requests.get() to return response codes.

```python
def check_sites(*urls):
    responses = []
    for url in urls:
        try:
            res = requests.get(url, timeout=5)
            responses.append((url, res.status_code))
        except requests.RequestException: # catch unreachable or invalid sites
            responses.append((url, None))
    return responses

results = check_sites(*sites)
print("Responses: ", results)
```

6.2. Use a list comprehension to print only URLs with status code 200.

```python
active_urls = [url for url, code in results if code == 200]

print("Active URLs:")
for site in active_urls:
    print(site)
```

6.3. Store results in a dictionary using a dictionary comprehension.

```python
status_dict = {url: code for url, code in results}
print("URL Status Dictionary:")
print(status_dict)
```

6.4. Create a generator expression that yields 'Active Site: <url>' for reachable domains

```python
active_sites_gen = ("Active Site: " + url for url, code in results if code == 200)

print("Generator Output:")
for msg in active_sites_gen:
    print(msg)
```
