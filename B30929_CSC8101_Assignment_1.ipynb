**Question 1: UBOS Multi-District Population & Growth Forecast**

1.1. Store the data for each district in a NumPy array.

```python
import numpy as np

kampala_data = np.array([1200, 1250, 1300, 1350, 1420, 1500, 1580, 1650, 1720, 1800])
wakiso_data = np.array([950, 1000, 1070, 1150, 1220, 1300, 1390, 1480, 1570, 1670])
gulu_data = np.array([320, 330, 345, 360, 375, 390, 410, 430, 455, 480])
```

1.2. Use the statistics module to compute mean, median, variance, and standard deviation for each district.

```python
import statistics
# Kampala
kampala_mean = statistics.mean(kampala_data)
kampala_median = statistics.median(kampala_data)
kampala_variance = np.var(kampala_data)
kampala_std_dev = np.std(kampala_data)
print("Kampala mean: ", kampala_mean, ", median: ", kampala_median, ", variance: ", kampala_variance, ", standard deviation: ", kampala_std_dev)
# Wakiso
wakiso_mean = statistics.mean(wakiso_data)
wakiso_median = statistics.median(wakiso_data)
wakiso_variance = np.var(wakiso_data)
wakiso_std_dev = np.std(wakiso_data)
print("Wakiso mean: ", wakiso_mean, ", median: ", wakiso_median, ", variance: ", wakiso_variance, ", standard deviation: ", wakiso_std_dev)
# Gulu
gulu_mean = statistics.mean(gulu_data)
gulu_median = statistics.median(gulu_data)
gulu_variance = np.var(gulu_data)
gulu_std_dev = np.std(gulu_data)
print("Gulu mean: ", gulu_mean, ", median: ", gulu_median, ", variance: ", gulu_variance, ", standard deviation: ", gulu_std_dev)
```

1.3. Generate a Fibonacci sequence of length 5 to project the next 5 years of growth.

```python
sequence = 5
fib_sequence = np.zeros(sequence, dtype=int)
fib_sequence[0] = 0
fib_sequence[1] = 1
for i in range((sequence - 2)):
    fib_sequence[i + 2] = fib_sequence[i + 1] + fib_sequence[i]
print("Fibonacci sequence of length 5: ", fib_sequence)

# Kampala
last_pop_size = kampala_data[-1]
last_growth_diff = last_pop_size - kampala_data[-2]
print("Kampala last growth difference: ", last_growth_diff)
projected_growth = np.array([fib_g * last_growth_diff for fib_g in fib_sequence])
kla_projected_data = last_pop_size + projected_growth
print("Kampala project the next 5 years of growth: ", kla_projected_data)
# Wakiso
last_pop_size = wakiso_data[-1]
last_growth_diff = last_pop_size - wakiso_data[-2]
print("Wakiso last growth difference: ", last_growth_diff)
projected_growth = np.array([fib_g * last_growth_diff for fib_g in fib_sequence])
wak_projected_data = last_pop_size + projected_growth
print("Wakiso project the next 5 years of growth: ", wak_projected_data)

# Gulu
last_pop_size = gulu_data[-1]
last_growth_diff = last_pop_size - gulu_data[-2]
print("Gulu last growth difference: ", last_growth_diff)
projected_growth = np.array([fib_g * last_growth_diff for fib_g in fib_sequence])
gulu_projected_data = last_pop_size + projected_growth
print("Gulu project the next 5 years of growth: ", gulu_projected_data)
```

1.4. Compare the variance of actual vs projected data.

```python
# Kampala
kla_projected_variance = statistics.variance(kla_projected_data)
print("Kampala actual variance: ", kampala_variance, ", projected variance: ", kla_projected_variance)
# Wakiso
wak_projected = statistics.variance(wak_projected_data)
print("Kampala actual variance: ", wak_projected, ", projected variance: ", kla_projected_data)
# Gulu
gulu_projected_variance = statistics.variance(gulu_projected_data)
print("Kampala actual variance: ", gulu_variance, ", projected variance: ", gulu_projected_variance)
```

1.5. Plot actual vs projected populations using Matplotlib.

```python
import matplotlib.pyplot as plt

years_actual = np.arange(10) + 1
years_projected = np.arange(5) + 11

plt.figure(figsize=(10, 6))

# Plot Actual Data
plt.plot(years_actual, kampala_data, 'o-', label='Kampala Actual (Years 1-10)', color='blue')
plt.plot(years_actual, wakiso_data, 'o-', label='Wakiso Actual (Years 1-10)', color='black')
plt.plot(years_actual, gulu_data, 'o-', label='Gulu Actual (Years 1-10)', color='red')
# Plot Projected Data
plt.plot(years_projected, kla_projected_data, 's--', label='Kampala Projected (Years 11-15)', color='orange')
plt.plot(years_projected, wak_projected_data, 's--', label='Wakiso Projected (Years 11-15)', color='grey')
plt.plot(years_projected, gulu_projected_data, 's--', label='Gulu Projected (Years 11-15)', color='green')

plt.title('Kampala, Wakiso, Gulu Population (Actual vs. Fibonacci Projection) ', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Population (in thousands)', fontsize=12)
plt.legend()
plt.xticks(np.concatenate([years_actual, years_projected]))
plt.grid(True, linestyle='--')
plt.show()
```

**Question 2: Solar Micro-Grid Simulation**
Equations: 3x + 2y = D1 and 4x + y = D2, where x = solar panels, y = batteries.

2.1. Prompt the user to enter daily demand values D1 and D2.

```python
from scipy.linalg import solve

eq_vector = np.array([[3, 2], [4, 1]])

solar_usage = []
battery_usage = []
day = 1
d1 = float(input("Enter Daily Demand D1 for Day " + str(day) + ": "))
d2 = float(input(f"Enter Daily Demand D2 for Day " + str(day) + ": "))

```

2.2. Use scipy.linalg.solve() to calculate energy from solar panels and batteries.

```python
demand_vector = np.array([d1, d2])
demand_sln = solve(eq_vector, demand_vector) # scipy.linalg.solve()
solar_usage.append(demand_sln[0]) # x = solar
battery_usage.append(demand_sln[1]) # y = batteries
```

2.3. Repeat for 7 days, storing results in a NumPy array.

```python
for day in range(2, 8):
  d1 = float(input("Enter Daily Demand D1 for Day " + str(day) + ": "))
  d2 = float(input("Enter Daily Demand D2 for Day " + str(day) + ": "))
  demand_vector = np.array([d1, d2])
  demand_sln = solve(eq_vector, demand_vector)
  solar_usage.append(demand_sln[0])
  battery_usage.append(demand_sln[1])
```

2.4. Use statistics to compute mean, variance, and standard deviation of solar vs
battery usage.

```python
solar_usage = np.array(solar_usage)
battery_usage = np.array(battery_usage)
print("Solar usage mean: ", statistics.mean(solar_usage), ", variance: ", np.var(solar_usage),", standard deviation: ", np.std(solar_usage))
print("Battery usage mean: ", statistics.mean(battery_usage), ", variance: ", np.var(battery_usage),", standard deviation: ", np.std(battery_usage))
```

2.5. Plot daily solar vs battery usage.

```python
days = np.arange(7) + 1

plt.figure(figsize=(10, 6))
plt.plot(days, solar_usage, 's-', label='Solar Panel Usage (x)', color='green')
plt.plot(days, battery_usage, 'o--', label='Battery Usage (y)', color='purple')

plt.title('Daily Solar vs. Battery Energy Usage ', fontsize=14)
plt.xlabel('Day', fontsize=12)
plt.ylabel('Energy Units (x or y)', fontsize=12)
plt.legend()
plt.xticks(days)
plt.grid(True, linestyle='--')
plt.show()
```

**Question 4: Weather & Agriculture Analysis**

4.1. Store rainfall data in a NumPy array.

```python
kampala_rainfall = np.array([120, 140, 180, 200, 220, 180, 90, 70, 60, 100, 110, 130])
```

4.2. Compute mean rainfall and classify months as 'Good for maize' (>150mm) or
'Drought risk' (<150mm).

```python
mean_rainfall = np.mean(kampala_rainfall)
print("Kampala mean rainfall: ", mean_rainfall)

print("Monthly Classification:")
month = 0
for rain in kampala_rainfall:
  month += 1
  classify = ""
  if rain > 150:
    classify = 'Good for maize'
  elif rain < 150:
    classify = 'Drought risk'
  print("Month ", month, "(", rain, "mm): ", classify)
```

4.3. Simulate Gulu rainfall with different values.

```python
gulu_rainfall_auto = []
for i in range(12):
  gulu_rainfall_auto.append(random.randint(100, 200))

gulu_rainfall = np.array(gulu_rainfall_auto)
print("Simulated Gulu rainfall data (mm): ", gulu_rainfall)
```

4.4. Compute cosine similarity of Kampala vs Gulu rainfall trends using math.cos().

```python
import math

dot_product = sum(a * b for a, b in zip(kampala_rainfall, gulu_rainfall))
magnitude_kla_rainfall = math.sqrt(sum(a**2 for a in kampala_rainfall))
magnitude_gulu_rainfall = math.sqrt(sum(b**2 for b in gulu_rainfall))

cosine_similarity = 0.0  # Handle cases where one or both lists are all zeros
if magnitude_kla_rainfall == 0 or magnitude_gulu_rainfall == 0:
  cosine_similarity = dot_product / (magnitude_kla_rainfall * magnitude_gulu_rainfall)

print("Cosine similarity of Kampala vs Gulu rainfall trends : ", cosine_similarity)


```

4.5. Plot both regions' rainfall in one chart.

```python
month_names = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

plt.figure(figsize=(12, 6))
plt.plot(month_names, kampala_rainfall, 'o-', label='Kampala Rainfall (mm)', color='blue', linewidth=2)
plt.plot(month_names, gulu_rainfall, 's--', label='Gulu Simulated Rainfall (mm)', color='brown', linewidth=2)

plt.title('Monthly Rainfall Comparison (Kampala vs Gulu) ', fontsize=14)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Rainfall (mm)', fontsize=12)
plt.axhline(150, color='red', linestyle=':', label='Maize Threshold (150mm)')
plt.legend()
plt.grid(True, axis='y', linestyle='--')
plt.show()
```

5.3. Use statistics to analyze mean, variance, and std deviation.

```python
mean_rev = statistics.mean(daily_revenues)
variance_rev = np.var(daily_revenues)
std_dev_rev = np.std(daily_revenues)
print("Revenue statistics mean: ", mean_rev, ", revenue: ", variance_rev, ", standard deviation: ", std_dev_rev)
```

5.4. Use scipy.linalg.solve to model a simple supply-demand system.

```python
# x = Supply, y = Demand
# 2*Supply + 1*Demand = Total Passengers
# 1*Supply + 3*Demand = Total Revenue
A_sd = np.array([[2, 1], [1, 3]])
for day in range(len(passenger_counts)):
  B_sd = np.array([passenger_counts[day], daily_revenues[day]]) # Using day data
  solution_sd = solve(A_sd, B_sd)
  supply_model = solution_sd[0]
  demand_model = solution_sd[1]
  print("Simple Supply-Demand Model (Day ", day,") supply (x):", supply_model, ", demand (y):", demand_model)
```

5.5. Forecast 11th day's revenue using average of last 3 days.

```python
last_3_passengers = passenger_counts[-3:]
avg_last_3_passengers = np.mean(last_3_passengers)
forecasted_revenue_day11 = avg_last_3_passengers * fare_per_passenger
print("Forecast last 3 days' passenger counts: ", last_3_passengers,
      ", average last 3 days' passengers: ", avg_last_3_passengers, ", forecasted revenue for day 11: ", forecasted_revenue_day11)
```

```python
all_days = np.arange(len(passenger_counts) + 1) + 1
actual_revenues_and_forecast = np.append(daily_revenues, forecasted_revenue_day11)

plt.figure(figsize=(10, 6))

# Plot actual revenues (Days 1-10)
plt.plot(all_days[:-1], daily_revenues, 'o-', label='Actual Revenue (Days 1-10)', color='darkblue')

# Plot the forecast point (Day 11)
plt.plot(all_days[-1], forecasted_revenue_day11, 'r*', markersize=12, label='Forecasted Revenue (Day 11)')

# Add a dashed line connecting the last actual point to the forecast
plt.plot([all_days[-2], all_days[-1]],
         [daily_revenues[-1], forecasted_revenue_day11],
         'r--', alpha=0.6)

plt.title('Pastport Revenue (Actual vs. Forecast) ', fontsize=14)
plt.xlabel('Day', fontsize=12)
plt.ylabel('Revenue (UGX)', fontsize=12)
plt.xticks(all_days)
plt.legend()
plt.grid(True, linestyle='--')
plt.show()
```
